## <span style='color:red'> 1] VISUALIZANDO LA TASA DE DESERCIÓN DE EMPLEADOS. </span>

```python
# Visualización para mostrar la pérdida de empleados en recuentos.
plt.figure(figsize=(17,6))
plt.subplot(1,2,1)
attrition_rate = employee_data["Attrition"].value_counts()
sns.barplot(x=attrition_rate.index,y=attrition_rate.values,palette=["#1d7874","#8B0000"])
plt.title("Número de empleados que han dejado la empresa",fontweight="black",size=20,pad=20)
for i, v in enumerate(attrition_rate.values):
    plt.text(i, v, v,ha="center", fontweight='black', fontsize=18)

#Visualización para mostrar la deserción de empleados en porcentaje.
plt.subplot(1,2,2)
plt.pie(attrition_rate, labels=["No","Yes"], autopct="%.2f%%", textprops={"fontweight":"black","size":15},
        colors = ["#1d7874","#AC1F29"],explode=[0,0.1],startangle=90)
center_circle = plt.Circle((0, 0), 0.3, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.title("Tasa de deserción de empleados",fontweight="black",size=20,pad=10)
plt.show()
```

### <font color=green>INFERENCIA:</font>

1. La tasa de deserción de empleados en esta organización es del 16,12 %.
2. Según expertos en Recursos Humanos, una tasa de deserción del 4 % al 6 % es normal en la organización.
3. Por lo tanto, podemos afirmar que la tasa de deserción de la organización se encuentra en un nivel peligroso.
4. Por lo tanto, la organización debería tomar medidas para reducirla.

## <span style='color:red'> 2] ANÁLISIS DE LA DESERCIÓN DE EMPLEADOS POR GÉNERO. </span>

```python
#Visualización para mostrar el total de empleados por género.
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
gender_attrition = employee_data["Gender"].value_counts()
plt.title("Distribución de empleados por género",fontweight="black",size=20)
plt.pie(gender_attrition, autopct="%.0f%%",labels=gender_attrition.index,textprops=({"fontweight":"black","size":20}),
        explode=[0,0.1],startangle=90,colors= ["#ffb563","#FFC0CB"])

#Visualización para mostrar la pérdida de empleados por género. 
plt.subplot(1,2,2)
new_df = employee_data[employee_data["Attrition"]=="Yes"]
value_1 = employee_data["Gender"].value_counts()
value_2 = new_df["Gender"].value_counts()
attrition_rate = np.floor((value_2/value_1)*100).values
sns.barplot(x=value_2.index, y=value_2.values,palette=["#D4A1E7","#E7A1A1"])
plt.title("Tasa de deserción de empleados por género",fontweight="black",size=20,pad=20)
for index,value in enumerate(value_2):
    plt.text(index,value,str(value)+" ("+str(int(attrition_rate[index]))+"% )",ha="center",va="bottom",
             size=15,fontweight="black")
plt.tight_layout()
plt.show()
```

### <font color=green>INFERENCIA:</font>

1. El número de empleados hombres en la organización representa una proporción mayor que el de empleados mujeres en más del 20 %.
2. Los empleados hombres se están yendo más de la organización que las empleadas.

## <span style='color:red'> 3] ANÁLISIS DE LA DESERCIÓN DE EMPLEADOS POR EDAD. </span>

```python
#Visualización para mostrar la distribución de empleados por edad.
plt.figure(figsize=(13.5,6))
plt.subplot(1,2,1)
sns.histplot(x="Age",hue="Attrition",data=employee_data,kde=True,palette=["#11264e","#6faea4"])
plt.title("Distribución de empleados X edad",fontweight="black",size=20,pad=10)

#Visualización para mostrar la distribución de empleados por edad y deserción.
plt.subplot(1,2,2)
sns.boxplot(x="Attrition",y="Age",data=employee_data,palette=["#D4A1E7","#6faea4"])
plt.title("Distribución de empleados edad X deserción",fontweight="black",size=20,pad=10)
plt.tight_layout()
plt.show()
```

### <font color=green>INFERENCIA:</font>

1. La mayoría de los empleados tienen entre 30 y 40 años.
2. Se observa claramente una tendencia: a medida que aumenta la edad, la deserción disminuye.
3. El diagrama de caja también muestra que la edad promedio de los empleados que dejaron la organización es menor que la de los empleados que aún trabajan en ella.
4. Los empleados jóvenes abandonan la empresa con mayor frecuencia que los de mayor edad.

## <span style='color:red'> 4] ANÁLISIS DE LA REDUCCIÓN DE EMPLEADOS POR VIAJES DE NEGOCIOS. </span>

```python
# Visualización para mostrar el total de empleados por viajes de negocios.
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
value_1 = employee_data["BusinessTravel"].value_counts()
plt.title("Empleados X viajes de negocios", fontweight="black", size=20, pad=20)
plt.pie(value_1.values, labels=value_1.index, autopct="%.1f%%",pctdistance=0.75,startangle=90,
        colors=['#E84040', '#E96060', '#E88181'],textprops={"fontweight":"black","size":15})
center_circle = plt.Circle((0, 0), 0.4, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# Visualización para mostrar la tasa de deserción por viajes de negocios.
plt.subplot(1,2,2)
new_df = employee_data[employee_data["Attrition"]=="Yes"]
value_2 = new_df["BusinessTravel"].value_counts()
attrition_rate = np.floor((value_2/value_1)*100).values
sns.barplot(x=value_2.index,y=value_2.values,palette=["#11264e","#6faea4","#FEE08B"])
plt.title("Tasa de deserción X viajes de negocios",fontweight="black",size=20,pad=20)
for index,value in enumerate(value_2):
    plt.text(index,value,str(value)+" ("+str(int(attrition_rate[index]))+"% )",ha="center",va="bottom",
             size=15,fontweight="black")
plt.tight_layout()
plt.show()
```

```python
# Define the bin edges for the groups
bin_edges = [0, 500, 1000, 1500]

# Define the labels for the groups
bin_labels = ['Low DailyRate', 'Average DailyRate', 'High DailyRate']

# Cut the DailyRate column into groups
employee_data['DailyRateGroup'] = pd.cut(employee_data['DailyRate'], bins=bin_edges, labels=bin_labels)
```

### <font color=green>INFERENCIA:</font>

1. Los empleados con una Tasa Diaria Promedio y una Tasa Diaria Alta son aproximadamente iguales.
2. Sin embargo, la tasa de deserción es muy alta entre los empleados con una Tasa Diaria Promedio en comparación con los empleados con una Tasa Diaria Alta.
3. La tasa de deserción también es alta entre los empleados con una Tasa Diaria Baja.
4. Los empleados que no reciben una Tasa Diaria Alta, en su mayoría, abandonan la organización.

## <span style='color:red'> 26] ANÁLISIS DE LA DESERCIÓN DE EMPLEADOS POR AÑOS CON EL GERENTE ACTUAL </span>

```python
# Define the bin edges for the groups
bin_edges = [0, 1, 5, 10, 20]

# Define the labels for the groups
bin_labels = ['0-1 years', '2-5 years', '5-10 years', "10+ years"]

# Cut the DailyRate column into groups
employee_data["YearsWithCurrManagerGroup"] = pd.cut(employee_data['YearsWithCurrManager'], bins=bin_edges, labels=bin_labels)
```

```python
#Visualization to show Total Employees by YearsWithCurrManagerGroup.
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
value_1 = employee_data["YearsWithCurrManagerGroup"].value_counts()
plt.title("Employees by YearsWithCurrManager", fontweight="black", size=20, pad=20)
plt.pie(value_1.values, labels=value_1.index, autopct="%.1f%%",pctdistance=0.75,startangle=90,
        colors= ['#FF8000', '#FF9933', '#FFB366', '#FFCC99'],textprops={"fontweight":"black","size":15})
center_circle = plt.Circle((0, 0), 0.4, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

#Visualization to show Attrition Rate by YearsWithCurrManagerGroup.
plt.subplot(1,2,2)
new_df = employee_data[employee_data["Attrition"]=="Yes"]
value_2 = new_df["YearsWithCurrManagerGroup"].value_counts()
attrition_rate = np.floor((value_2/value_1)*100).values
sns.barplot(x=value_2.index.tolist(), y=value_2.values,palette=["#11264e","#6faea4","#FEE08B","#D4A1E7","#E7A1A1"])
plt.title("Attrition Rate by YearsWithCurrManager",fontweight="black",size=20,pad=20)
for index,value in enumerate(value_2):
    plt.text(index,value,str(value)+" ("+str(int(attrition_rate[index]))+"%)",ha="center",va="bottom",
             size=15,fontweight="black")
plt.tight_layout()
plt.show()
```
