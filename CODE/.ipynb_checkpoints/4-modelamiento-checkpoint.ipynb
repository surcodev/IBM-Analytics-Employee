<div style="text-align: center; background-color: yellow; padding: 10px;">
    <h2 style="font-weight: bold;">PROCESAMIENTO DE DATOS</h2>
</div>

````python
# Transform categorical data into dummies
dummy_col = [column for column in employee_data.drop('Attrition', axis=1).columns if employee_data[column].nunique() < 20]
data = pd.get_dummies(employee_data, columns=dummy_col, drop_first=True, dtype='uint8')
data.info()
````

````python
print(data.shape)

# Eliminar caracteristicas duplicadas
data = data.T.drop_duplicates()
data = data.T

# Eliminar filas duplicadas
data.drop_duplicates(inplace=True)

print(data.shape)
````

````python
data.drop('Attrition', axis=1).corrwith(data.Attrition).sort_values().plot(kind='barh', figsize=(10, 30))
````

En el gráfico que generaste, cada barra representa el **coeficiente de correlación** entre una variable del dataset y la variable objetivo **Attrition** (deserción de empleados, codificada como 1 = Sí, 0 = No).

Te lo explico por partes:

---

### **1. Eje horizontal (valores de correlación)**

* **Positivo**: valores hacia la derecha (mayor que 0) indican que **aumentar esa variable se asocia con más probabilidad de Attrition**.
  Ejemplo: `OverTime_Yes` tiene la correlación más alta (\~0.25), lo que significa que quienes hacen horas extra tienen más tendencia a irse de la empresa.
* **Negativo**: valores hacia la izquierda (menor que 0) indican que **aumentar esa variable se asocia con menos probabilidad de Attrition**.
  Ejemplo: `TotalWorkingYears` (\~ -0.17) sugiere que empleados con más años trabajados tienden a quedarse.

---

### **2. Eje vertical (variables)**

* Cada nombre es una característica del dataset.
* Muchas variables están codificadas en formato **dummy** (ejemplo: `JobRole_Sales Representative` o `JobInvolvement_Low`).

---

### **3. Patrones visibles**

* **Mayor correlación positiva con Attrition**:

  * `OverTime_Yes`: muy relacionado con la deserción.
  * `MaritalStatus_Single`: empleados solteros muestran más propensión a irse.
  * Bajo `EnvironmentSatisfaction` o bajo `JobInvolvement`: más probabilidad de dejar la empresa.
  * `BusinessTravel_Travel_Frequently`: viajar mucho se asocia con mayor deserción.

* **Mayor correlación negativa con Attrition**:

  * `TotalWorkingYears`, `Age`, `YearsAtCompany`: empleados con más experiencia y antigüedad tienden a quedarse.
  * Puestos como `JobRole_Research Director`, `JobRole_Manager` o `Manufacturing Director` están asociados a menos deserción.
  * Estado civil `Married` también tiene correlación negativa.

---

### **4. Interpretación práctica**

Este gráfico sirve para:

* Identificar **factores de riesgo** de deserción (correlaciones positivas).
* Reconocer **factores protectores** contra la deserción (correlaciones negativas).
* Priorizar variables para un modelo predictivo o para políticas de retención.

````python
X = data.drop('Attrition', axis=1)
y = data.Attrition

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,
                                                    stratify=y)

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
X_std = scaler.transform(X)
````

````python
def feature_imp(df, model):
    fi = pd.DataFrame()
    fi["feature"] = df.columns
    fi["importance"] = model.feature_importances_
    return fi.sort_values(by="importance", ascending=False)
````

````python
y_test.value_counts()[0] / y_test.shape[0]
````

````python
stay = (y_train.value_counts()[0] / y_train.shape)[0]
leave = (y_train.value_counts()[1] / y_train.shape)[0]

print("===============ENTRENAMIENTO=================")
print(f"Tasa de permanencia: {stay * 100:.2f}%")
print(f"Tasa de abandono: {leave * 100 :.2f}%")

stay = (y_test.value_counts()[0] / y_test.shape)[0]
leave = (y_test.value_counts()[1] / y_test.shape)[0]

print("===============PRUEBA=================")
print(f"Tasa de permanencia: {stay * 100:.2f}%")
print(f"tasa de abandono: {leave * 100 :.2f}%")
````

````python
def evaluate(model, X_train, X_test, y_train, y_test):
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    print("RESULTADO DEL ENTRENAMIENTO: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_train, y_train_pred, output_dict=True))
    print(f"CONFUSION MATRIX:\n{confusion_matrix(y_train, y_train_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_train, y_train_pred):.4f}")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")

    print("RESULTADO DE LAS PRUEBAS: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_test, y_test_pred, output_dict=True))
    print(f"CONFUSION MATRIX:\n{confusion_matrix(y_test, y_test_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_test, y_test_pred):.4f}")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")
````

<div style="text-align: center; background-color: yellow; padding: 10px;">
    <h2 style="font-weight: bold;">REGRESION LOGISTICA</h2>
</div>

````python
lr_clf = LogisticRegression(solver='liblinear', penalty='l1')
lr_clf.fit(X_train_std, y_train)

evaluate(lr_clf, X_train_std, X_test_std, y_train, y_test)
````

Vamos a desmenuzar esos resultados paso a paso:

---

## **1. ENTRENAMIENTO**

### **Matriz de confusión**

```
[[847  16]
 [ 59 107]]
```

* **TN (847)**: predijo correctamente que el empleado *no se va* (clase 0).
* **FP (16)**: predijo que se iba, pero realmente no se iba.
* **FN (59)**: predijo que no se iba, pero sí se fue.
* **TP (107)**: predijo correctamente que se iba (clase 1).

📌 **Interpretación**:

* Muy buen desempeño en detectar clase 0 (no se va).
* Clase 1 (sí se va) está más difícil de predecir: 59 falsos negativos.

---

### **Accuracy entrenamiento**: `0.9271` (92.71%)

Muy alto, pero ojo: el dataset seguramente está desbalanceado (muchos más 0 que 1).

---

### **Reporte de clasificación**

* **Precision clase 0**: 0.9349 → Cuando predice “no se va”, acierta el 93.5%.
* **Recall clase 0**: 0.9815 → Detecta el 98% de los que realmente no se van.
* **Precision clase 1**: 0.8699 → Cuando predice “sí se va”, acierta el 87%.
* **Recall clase 1**: 0.6446 → Solo detecta \~64% de los que realmente se van (se le escapan varios).

---

## **2. PRUEBAS**

### **Matriz de confusión**

```
[[351  19]
 [ 41  30]]
```

* **TN (351)**: predijo correctamente que no se va.
* **FP (19)**: falso positivo (predijo que se iba, pero no).
* **FN (41)**: falso negativo (predijo que no se iba, pero sí).
* **TP (30)**: predijo correctamente que se iba.

📌 **Interpretación**:

* Baja detección de los que sí se van (TP=30 vs FN=41).
* Clase 1 sigue siendo el punto débil.

---

### **Accuracy pruebas**: `0.8639` (86.39%)

* El modelo baja su rendimiento en datos nuevos (de 92.71% a 86.39%).
* Esto puede ser señal de **overfitting** (memoriza más de lo que generaliza).

---

### **Reporte de clasificación**

* **Precision clase 0**: 0.8954
* **Recall clase 0**: 0.9486
* **Precision clase 1**: 0.6122 → cuando predice que se va, solo acierta un 61%.
* **Recall clase 1**: 0.4225 → detecta menos de la mitad de los que realmente se van.
* **F1-score clase 1**: 0.50 → indica que el rendimiento para detectar renuncias es limitado.

---

## **Conclusiones**

1. El modelo es muy bueno prediciendo empleados que **se quedan** (clase 0).
2. Es mucho menos preciso para detectar **quién se va** (clase 1).
3. Probablemente el dataset está desbalanceado, lo que favorece la clase 0.
4. En pruebas, baja su desempeño → posible **overfitting**.
5. Para mejorar detección de clase 1 se podría:

   * Balancear los datos (SMOTE, undersampling, oversampling).
   * Ajustar el umbral de decisión del clasificador.
   * Usar métricas como F1-score o Recall como objetivo, no solo accuracy.


````python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g--", label="Recall")
    plt.xlabel("Threshold")
    plt.legend(loc="upper left")
    plt.title("Precision/Recall Tradeoff")
    

def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], "k--")
    plt.axis([0, 1, 0, 1])
    plt.xlabel('Tasa de falsos positivos')
    plt.ylabel('Tasa de verdaderos positivos')
    plt.title('Curva ROC')
    
    
precisions, recalls, thresholds = precision_recall_curve(y_test, lr_clf.predict(X_test_std))
plt.figure(figsize=(14, 25))
plt.subplot(4, 2, 1)
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)

plt.subplot(4, 2, 2)
plt.plot(precisions, recalls)
plt.xlabel("Precision")
plt.ylabel("Recall")
plt.title("PR Curve: precisions/recalls tradeoff");

plt.subplot(4, 2, 3)
fpr, tpr, thresholds = roc_curve(y_test, lr_clf.predict(X_test_std))
plot_roc_curve(fpr, tpr)
````

````python
scores_dict = {
    'Logistic Regression': {
        'Train': roc_auc_score(y_train, lr_clf.predict(X_train)),
        'Test': roc_auc_score(y_test, lr_clf.predict(X_test)),
    },
}
````

    <h2 style="font-weight: bold;">RANDOM FOREST CLASSIFIER</h2>
</div>

````python
rf_clf = RandomForestClassifier(n_estimators=100, bootstrap=False,
#                                      class_weight={0:stay, 1:leave}
                                    )
rf_clf.fit(X_train, y_train)
evaluate(rf_clf, X_train, X_test, y_train, y_test)
````

````python
param_grid = dict(
    n_estimators= [100, 500, 900], 
    max_features= ['auto', 'sqrt'],
    max_depth= [2, 3, 5, 10, 15, None], 
    min_samples_split= [2, 5, 10],
    min_samples_leaf= [1, 2, 4], 
    bootstrap= [True, False]
)

rf_clf = RandomForestClassifier(random_state=42)
search = GridSearchCV(rf_clf, param_grid=param_grid, scoring='roc_auc', cv=5, verbose=1, n_jobs=-1)
search.fit(X_train, y_train)

rf_clf = RandomForestClassifier(**search.best_params_, random_state=42)
rf_clf.fit(X_train, y_train)
evaluate(rf_clf, X_train, X_test, y_train, y_test)
````

````python
precisions, recalls, thresholds = precision_recall_curve(y_test, rf_clf.predict(X_test))
plt.figure(figsize=(14, 25))
plt.subplot(4, 2, 1)
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)

plt.subplot(4, 2, 2)
plt.plot(precisions, recalls)
plt.xlabel("Precision")
plt.ylabel("Recall")
plt.title("PR Curve: precisions/recalls tradeoff");

plt.subplot(4, 2, 3)
fpr, tpr, thresholds = roc_curve(y_test, rf_clf.predict(X_test))
plot_roc_curve(fpr, tpr)
````

````python
scores_dict['Random Forest'] = {
        'Train': roc_auc_score(y_train, rf_clf.predict(X_train)),
        'Test': roc_auc_score(y_test, rf_clf.predict(X_test)),
    }
````

````python
plt.figure(figsize=(10, 8))

# Iterate through each model
for model_name, model in ml_models.items():
    if hasattr(model, "predict_proba"):  # Check if the model has predict_proba method
        # Predict probabilities for positive class
        y_pred_prob = model.predict_proba(X_test)[:, 1]
    else:
        # For models without predict_proba, use decision_function or predict
        y_pred_prob = model.decision_function(X_test) if hasattr(model, "decision_function") else model.predict(X_test)
    
    # Compute ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    
    # Compute AUC score
    auc_score = roc_auc_score(y_test, y_pred_prob)
    
    # Plot ROC curve
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc_score:.3f})')

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random guessing
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
````
