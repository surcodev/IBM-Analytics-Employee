```python
# Library for Data Manipulation
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)

# Library for Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white",font_scale=1.5)
sns.set(rc={"axes.facecolor":"#FFFAF0","figure.facecolor":"#FFFAF0"})
sns.set_context("poster",font_scale = .7)

# Library to perform Statistical Analysis.
from scipy import stats
from scipy.stats import chi2
from scipy.stats import chi2_contingency

# Library for Ignore the warnings
import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')
```

```python
employee_data = pd.read_csv(r'IBM-HR-Analytics-Employee-Attrition-and-Performance-Revised.csv')
```

## <span style='color:blue'> 1] ANOVA TEST </span>
<span style='color:red'> ANALIZAR LA IMPORTANCIA DE LAS CARACTERSTICAS NUMRICAS EN LA DESERCION DE EMPLEADOS </span>

### <font color=green>INFERENCIA:</font>

La prueba ANOVA se utiliza para analizar si las medias de una <b>caracterstica numrica</b> difieren significativamente entre los grupos de una <b>caracterstica categrica</b>, en este caso, el estado de desercin de los empleados (Attrition: "Yes" o "No").
La prueba ANOVA devuelve dos valores estadsticos: F-score y p-value.

### <font color=black>Nota:</font>

- Un F-score ms alto indica una asociacin ms fuerte entre la caracterstica numrica (variable independiente) y el estado de desercin (variable dependiente categrica).  
- Si el p-value est por debajo del nivel de significancia elegido (por ejemplo, p < 0.05), podemos concluir que existe una diferencia estadsticamente significativa en las medias y rechazar la hiptesis nula.

```python
num_cols = employee_data.select_dtypes(np.number).columns
```

```python
new_df = employee_data.copy()

new_df["Attrition"] = new_df["Attrition"].replace({"No":0,"Yes":1})
```

```python
f_scores = {}
p_values = {}

for column in num_cols:
    f_score, p_value = stats.f_oneway(new_df[column],new_df["Attrition"])
    
    f_scores[column] = f_score
    p_values[column] = p_value
```

## <span style='color:blue'> 2] VISUALIZAR LA F_SCORE DE LA PRUEBA ANOVA </span>
<span style='color:red'> DE CADA CARACTERSTICA NUMRICA. </span>

```python
plt.figure(figsize=(15,6))
keys = list(f_scores.keys())
values = list(f_scores.values())

sns.barplot(x=keys, y=values)
plt.title("Comparacion de F_scores en la prueba ANOVA",fontweight="black",size=20,pad=15)
plt.xticks(rotation=90)

for index,value in enumerate(values):
    plt.text(index,value,int(value), ha="center", va="bottom",fontweight="black",size=15)
plt.show()
```

## <span style='color:blue'> 3] COMPARANDO F_SCORE Y P_VALUE DE LA PRUEBA ANOVA. </span>

```python
test_df = pd.DataFrame({"Features":keys,"F_Score":values})
test_df["P_value"] = [format(p, '.100f') for p in list(p_values.values())]
```

```python
test_df
```

## <span style='color:blue'> 4] REALIZAR PRIEBA CHI-2 </span>
<span style='color:red'> ANALIZAR LA IMPORTANCIA DE LAS CARACTERSTICAS CATEGRICAS EN LA DESERCIN DE EMPLEADOS </span>

```python
cat_cols = employee_data.select_dtypes(include="object").columns.tolist()
cat_cols.remove("Attrition")
```

```python
chi2_statistic = {}
p_values = {}

# Perform chi-square test for each column
for col in cat_cols:
    contingency_table = pd.crosstab(employee_data[col], employee_data['Attrition'])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    chi2_statistic[col] = chi2
    p_values[col] = p_value
```

## <span style='color:blue'> 5] VISUALIZAR LOS VALORES ESTADSTICOS DE CHI-CUADRADO </span>
<span style='color:red'> DE CARACTERSTICAS CATEGRICAS </span>

```python
columns = list(chi2_statistic.keys())
values = list(chi2_statistic.values())

plt.figure(figsize=(16,6))
sns.barplot(x=columns, y=values)
plt.xticks(rotation=90)
plt.title("Valor estadstico Chi2 de cada columna categrica",fontweight="black",size=20,pad=15)
for index,value in enumerate(values):
    plt.text(index,value,round(value,2),ha="center",va="bottom",fontweight="black",size=15)

plt.show()
```

## <span style='color:blue'> 6] COMPARACIN DEL ESTADISTICO DE CHI^2 Y EL P_SCORE DE LA PRUEBA DE CHI^2. </span>

```python
test_df = pd.DataFrame({"Features":columns,"Chi_2 Statistic":values})
test_df["P_value"] =  [format(p, '.20f') for p in list(p_values.values())]
```

```python
test_df
```
