**Attition**
- Renuncia voluntaria
- Despido
- Desgaste
- Jubilación

Attrition = "Yes" → El empleado ya no trabaja en la empresa.
Attrition = "No" → El empleado aún trabaja en la empresa.

**¿Por qué es importante?**
Las empresas analizan la tasa de attrition para:
- Identificar causas de renuncias
- Prevenir la fuga de talento
- Mejorar el clima laboral
- Reducir costos por rotación
